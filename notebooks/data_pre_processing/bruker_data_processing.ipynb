```python
dat = pd.read_excel('data/rawdata/bruker/PROJ0074_100FDR_GrapePI_Report.xlsx', header=None, skiprows=2)
# dat = pd.read_csv('data/rawdata/bruker/PROJ0074_CGU3_NoFilter_100FDR_Report.tsv', sep='\t')
dat
```

```python
# data is in excel format with two level column, which is not convenient to work with
# the following code read the Cscore from three PROJ0074_SNU16 replicates
dat = dat.iloc[:, [4, 5, 20, 21, 22, 23, 24, 25]]
dat.columns = ['accession', 'gene_symbol', 'MKN45_run1', 'MKN45_run2', 'MKN45_run3', 'SNU16_run1', 'SNU16_run2', 'SNU16_run3']
dat
```

```python
# Ideally you should keep Cscore for all proteins including decoy proteins
# however, in this dataset we have some proteins and all decoy with missing Cscore
# so we need to remove them. Missing value for decoy proteins is not a problem (can still use EM algorithm to calculate protein probability)
# but too many missing values for target proteins is a problem (may lead into incorrect null distribution estimation)

# check number of missing values
print(dat.isnull().sum())

# # check how many accession has "decoy" in it
# dat['accession'].str.contains('Decoy').sum()

# fortunately, not many missing values for target proteins
```

```python
# remove missing values
dat = dat.dropna()
```

```python
# some proteins with almost identical amino acid sequence are not distinguished and reported in the same row
dat.loc[:, 'accession'] = dat['accession'].str.split(';')
dat = dat.explode('accession')
dat
```

```python
# plot histogram of protein probability
dat['MKN45_run1'].hist(bins=100)
```

```python
# correlation between
import seaborn as sns
# calculate spearman correlation
import scipy.stats as stats
stats.spearmanr(dat['MKN45_run2'], dat['MKN45_run1'])
```

```python
# load swissprot data
swissprot = pd.read_csv('data/miscellaneous/uniprotkb_proteome_UP000005640_AND_revi_2023_10_05.tsv', sep='\t')
swissprot.rename(columns={'Entry': 'accession', 'Gene Names (primary)': 'gene_symbol'}, inplace=True)
```

```python
swissprot
```

```python
# check how many is in swissprot
dat['accession'].isin(swissprot['accession']).sum()
```

```python

# only keep those in swissprot
dat[dat['accession'].isin(swissprot['accession'])]
# add those not in swissprot with value 
```

```python
dat_swissprot = dat.merge(swissprot, on=['accession', 'gene_symbol'], how='right')
# fill NA with 0
dat_swissprot['MKN45_run1'].fillna(1, inplace=True)
dat_swissprot
```
